```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("cleaned_imdb.csv")
```

```python
steps = pd.read_csv("data_prep_recipe.csv")
```

```python
num_cols = steps.loc[steps["datatype"]==1,"column"].tolist()
```

```python
cat_cols=[]
for i in range(0,df.shape[1]):
    if df.columns[i] not in num_cols:
        cat_cols.append(df.columns[i])
```

```python
from matplotlib.pyplot import figure
import numpy as np
```

```python
plt.scatter(df["director_facebook_likes"],df["imdb_score"])
plt.show()
df["director_facebook_likes"].min()
```

```python
for col in num_cols:
    figure(figsize=(18, 8), dpi=80)
    min_offset = 0


    if(df[col].min() < 1):
        min_offset = (1 - df[col].min()) + .1
    cor = round(np.corrcoef(np.log(df[col]+min_offset),df["imdb_score"])[1][0],2)
    print(cor)
    plt.scatter(np.log(df[col]+min_offset),df["imdb_score"],label=col)
    plt.title(col+": "+str(cor))
    plt.legend()
    plt.show()
    df[col+"_log"] = np.log(df[col]+min_offset)
```

```python
np.corrcoef(np.log(df["director_facebook_likes"]+1.1),df["imdb_score"])
```
