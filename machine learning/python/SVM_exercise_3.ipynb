```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from math import sqrt
import matplotlib
matplotlib.rcParams["figure.figsize"] = [6,6]
```

```python
# generate linearly seperable data
def gen_circular_data(N):
    x = np.random.rand(N)
    y = np.random.rand(N)
    
    df = pd.DataFrame({"x":x,"y":y})
    center1 = (.3,.75)
    center2 = (.75,.3)
    radius = .2
    def check_line_side(x):
        if sqrt((x["x"] - center1[0])*(x["x"] - center1[0]) + (x["y"] - center1[1])*(x["y"] - center1[1])) < radius:
            return 2
        elif sqrt((x["x"] - center2[0])*(x["x"] - center2[0]) + (x["y"] - center2[1])*(x["y"] - center2[1])) < radius:
            return 1
        else:
            return 0
    y = df.apply(check_line_side,axis=1)
    return df, y
```

```python
X, y = gen_circular_data(1000)
```

```python
plt.scatter(X.x,X.y,c=y)
plt.show()
```

```python
#clf = SVC(kernel="rbf",gamma=5)
clf = SVC(kernel="poly",degree=3,coef0=1)
```

```python
clf.fit(X,y)
```

```python
pred = clf.predict(X)
```

```python
plt.scatter(X.x,X.y,c=pred)
plt.show()
```
