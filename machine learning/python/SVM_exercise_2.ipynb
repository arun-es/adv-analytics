```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.svm import SVC
```

```python
# generate linearly seperable data
def gen_non_linear_data(N):
    x = np.random.rand(N)
    y = np.random.rand(N)
    
    df = pd.DataFrame({"x":x,"y":y})

    def check_line_side(x):
        if(np.sin(x["x"]*10 - .5)/5+.5< x["y"]):
            return 1
        else:
            return 0
    y = df.apply(check_line_side,axis=1)
    return df, y
```

```python
X, y = gen_non_linear_data(1000)
```

```python
plt.scatter(X.x,X.y,c=y)
plt.show()
```

```python
# Building a poly svm
clf = SVC(kernel="poly",degree=5,coef0=1)
clf = SVC(kernel="rbf",gamma=100)
```

```python
clf.fit(X,y)
```

```python
pred = clf.predict(X)
```

```python
plt.scatter(X.x,X.y,c=pred)
plt.show()
```
