```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.svm import SVC
```

```python
# generate linearly seperable data
def gen_linear_data(N):
    x = np.random.rand(N)
    y = np.random.rand(N)
    
    df = pd.DataFrame({"x":x,"y":y})

    def check_line_side(x):
        if(x["x"] + np.random.rand()*.5 < x["y"]  + np.random.rand()*.1):
            return 1
        else:
            return 0
    y = df.apply(check_line_side,axis=1)
    return df, y
```

```python
X, y = gen_linear_data(1000)
```

```python
print(X.head())
```

```python
plt.scatter(X.x,X.y,c=y)
plt.show()
```

```python
# Building a linear svm
clf = SVC(kernel="linear")
```

```python
clf.fit(X,y)
```

```python
pred = clf.predict(X)
```

```python
plt.scatter(X.x,X.y,c=pred)
plt.show()
```
